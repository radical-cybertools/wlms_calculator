```python
loc = '../../data/validation'
fontsize=18
folds = ['set-%s'%x for x in range(1,10)]
```

```python
def parse_uids(uid):
    new_uid = int(uid.split('.')[-1])
    return new_uid
```

```python
dfs = list()
for fold in folds:
    for x in glob('%s/%s/profile.*'%(loc, fold)):
        json_data = ru.read_json(x)
        df = pd.DataFrame.from_dict(json_data[json_data.keys()[0]])
        df['core'] = df['core'].apply(parse_uids)
        df['task'] = df['task'].apply(parse_uids)
        df['perf'] = df['ops']/df['exec_time']
        dfs.append(df)
    
print len(dfs)
```

```python
def plotter(df):
    
    fig, axes = fig_setup(2,2,(24,16), (0.35,0.35))
    
    ax = df.plot(ax=axes[0][0], y=['ops'], ylim=(900,1200), title='Number of operations to be performed by each task')
    ticks = [x for x in range(len(df['task']))]
    ax.set_xticks(ticks)
    ax.set_xticklabels(df['task'].tolist(), rotation=90)
    ax.set_xlabel('Task ID')
    ax.set_ylabel('No. of operations')

    new_df = df.drop_duplicates('core')
    ax = new_df.plot(ax=axes[0][1], y=['perf'], ylim=(10,90), title='Number of operations performed per second by each core')
    ticks = [x for x in range(len(new_df['core']))]
    ax.set_xticks(ticks)
    ax.set_xticklabels(new_df['core'].tolist(), rotation=90)
    ax.set_xlabel('Core ID')
    ax.set_ylabel('No. of operations per second')  
    
    ax = df.plot(ax=axes[1][0], y=['start_time','end_time'], ylim=(-10,90), title='Actual execution profile (WLMS)')
    ticks = [x for x in range(len(df['task']))]
    ax.set_xticks(ticks)
    ax.set_xticklabels(df['task'].tolist(), rotation=90)
    ax.set_xlabel('Task ID')
    ax.set_ylabel('Time (seconds)')

    df_manual = df[['core','ops','task','perf']]
    num_cores = len(set(df_manual['core'].tolist()))
    num_tasks = len(df_manual)
    prev = dict()
    for core in df['core']:
        prev[core] = 0
        
    for t in range(num_tasks):
        df_manual.loc[t,'start_time'] = prev[df.loc[t,'core']]
        df_manual.loc[t,'end_time'] = prev[df.loc[t,'core']] + df_manual.loc[t,'ops'] / df_manual.loc[t,'perf']
        prev[df.loc[t,'core']] = df_manual.loc[t,'ops'] / df_manual.loc[t,'perf']
    
    
#     ax = df_manual.plot(ax=axes[1][1], y=['start_time','end_time'], ylim=(-10,90), title='Expected execution profile (Manual)')
#     ticks = [x for x in range(len(df['task']))]
#     ax.set_xticks(ticks)
#     ax.set_xticklabels(df['task'].tolist(), rotation=90)
#     ax.set_xlabel('Task ID')
#     ax.set_ylabel('Time (seconds)')

    
    df_diff = pd.DataFrame(columns=['diff_start_time','diff_end_time', 'task'])
    df_diff['task'] = df['task']
    df_diff['diff_start_time'] = df['start_time'] - df_manual['start_time']
    df_diff['diff_end_time'] = df['end_time'] - df_manual['end_time']
    ax = df_diff.plot(ax=axes[1][1], y= ['diff_start_time','diff_end_time'], title='Difference b/w actual and expected calculation')
    ticks = [x for x in range(len(df_diff['task']))]
    ax.set_xticks(ticks)
    ax.set_xticklabels(df_diff['task'].tolist(), rotation=90)
    ax.set_xlabel('Task ID')
    ax.set_ylabel('Time (seconds)')
    
    
    return fig
```

```python
def plotter_3d(df):
    
    def parse_uids_2(uid):
        if isinstance(uid, int):
            return uid
        uid = int(uid.split('.')[-1])
        return uid
        
    df['task'] = df['task'].apply(parse_uids_2)
    df['core'] = df['core'].apply(parse_uids_2)
    
    fig = plt.figure(figsize=(24,12))
    ax = fig.add_subplot(111, projection='3d')
    for task in df['task']:
        row = df[df['task']==task]
        x = row['core']
        y = task
        bottom=row['start_time']
        top = row['end_time'] - bottom        
        width=0.1
        depth=0.1
        ax.bar3d(x, y, bottom, width, depth, top, color='blue', shade=True)
    
    ax.set_xlabel('Core ID', labelpad=20)
    ax.set_xticks(df['core'].tolist())
    ax.set_ylabel('Task ID', labelpad=25)
    ax.set_yticks(df['task'].tolist())
    ax.set_zlabel('Time (secs)', labelpad=20)
#     ax.set_zlim(0,50)
    return fig
```

```python
df = dfs[0]
fig = plotter(df)
title = 'Validation of execution time given a homogoneous workload and resource'
fig.suptitle(title, y=0.95)
# txt = 'Config used:'+ \
#         '\nPrimitive order = [task_selection, resource_selection, spatial_binding, temporal_binding]' + \
#         '\nCriteria = [task_selector: all, resource_selector: all, binder: l2f]' + \
#         '\nWorkload = [num_tasks=4, ops_dist=uniform,  dist_mean=1024, dist_var=0] (1 workload of this config submitted every 5 seconds for a total of 2 workloads)' + \
#         '\nResource = [num_cores=4, perf_dist=uniform, dist_mean=32, temporal_var=0, spatial_var=0]' 
# x = fig.text(0.1,-0.1, txt)
fig.savefig('plots/hom_wl_hom_res_1.png', dpi=fig.dpi)
```
